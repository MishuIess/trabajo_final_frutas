```python
# IMPORTAR LIBRERÍAS

import sqlite3
import pandas as pd
from faker import Faker
import random
```

```python
# Inicializar Faker
fake = Faker()

# Conectar o crear base de datos SQLite
conexion = sqlite3.connect("personas_frutas.sqlite")

# Crear lista de datos
datos = []
```

```python
# GENERAR DATOS

datos = []

for i in range(1, 5001):

    datos.append({

        "SurveyID": i,

        "Nombre": fake.name(),

        "Edad": random.randint(18, 70),

        "Ciudad": fake.city(),

        "FrutaFavorita": random.choice([
            "Manzana",
            "Banana",
            "Naranja",
            "Pera",
            "Uva",
            "Sandía",
            "Melón",
            "Piña",
            "Mango",
            "Fresa"
        ]),

        "ConsumoSemanal": random.randint(1, 20)

    })

# CREAR DATAFRAME

df_personas = pd.DataFrame(datos)
```

```python
# GUARDAR EN SQLITE

df_personas.to_sql(
    "personas_consumo_frutas",
    conexion,
    if_exists="replace",
    index=False
)
```

```python
# CONSULTAR DATOS

query = """
SELECT *
FROM personas_consumo_frutas
LIMIT 5;
"""

resultado = pd.read_sql_query(
    query,
    conexion
)

print(resultado)
```

```python
# VERIFICAR TABLAS SQLITE

query_tablas = """
SELECT name
FROM sqlite_master
WHERE type='table';
"""

tablas = pd.read_sql_query(
    query_tablas,
    conexion
)

print("\nTABLAS EXISTENTES")
print(tablas)
```

```python
# Cerrar conexión
conexion.close()

print("Base de datos SQLite creada correctamente")
```
