```python
# IMPORTAR LIBRERÍAS

from faker import Faker
import pandas as pd
import random
```

```python
# Inicializar Faker
fake = Faker()

# Lista de frutas
frutas = [
    "Manzana", "Banana", "Naranja", "Pera", "Uva",
    "Sandía", "Melón", "Piña", "Mango", "Fresa"
]

# Lista para guardar datos
datos = []
```

```python
# Generar 5000 registros
for i in range(1, 5001):
    datos.append({
        "ID": i,
        "Fruta": random.choice(frutas),
        "Cantidad": random.randint(1, 100),
        "Precio": round(random.uniform(0.5, 15.0), 2),
        "Fecha": fake.date_between(start_date='-2y', end_date='today')
    })

# Crear DataFrame
df = pd.DataFrame(datos)
```

```python
# Guardar en CSV
df.to_csv("frutas_5000_datos.csv", index=False)
```
